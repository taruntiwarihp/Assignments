1. Display all the information respective to gender

2. Find all the male who is consuming alcohal more than 5 years and has diabetes result positive.

```python
df=hc_df[(hc_df['Gender']=='male') & (hc_df['Duration of alcohol consumption(years)'] > 5) & (hc_df['Diabetes Result']=='YES')]
df.head()
```

3. Find where more people consume alcohol in rural or urban?

```python
rural_cons = hc_df[hc_df['Place(location where the patient lives)']=='rural']['Quantity of alcohol consumption (quarters/day)'].values.sum()
urban_cons = hc_df[hc_df['Place(location where the patient lives)']=='urban']['Quantity of alcohol consumption (quarters/day)'].values.sum()
if rural_cons > urban_cons:
    print("Rural area has more consumtion")
else:
    print("Urban area has more consumtion")
```

4. Find average age of male and female who consume alcohol.

```python
df= hc_df[hc_df['Quantity of alcohol consumption (quarters/day)']>0]
df.groupby('Gender')['Age'].mean()
```

8. If gender is male and age is less than 35 then change diabetes result to NO .

```python
hc_df['Diabetes Result'] = hc_df.apply(lambda x: 'NO' if (x['Gender']== 'male') and (x['Age'] < 35) else x['Diabetes Result'],axis=1) 
```

9. Find all record where Place is missing.

```python
hc_df[hc_df['Place(location where the patient lives)'].isnull()]
```

10. Change place to urban if gender is female else rural.

```python
def change_place(d):
    if 'female' in d["Gender"]:
        return 'urban'
    else:
        return 'rural'
hc_df['Place(location where the patient lives)']=hc_df.apply(change_place,axis=1)
```
