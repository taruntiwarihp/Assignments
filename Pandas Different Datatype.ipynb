```python
import pandas as pd
import xml.etree.ElementTree as etree
pd.__version__
```

```python
#Read student.xml file
student=etree.parse("student.xml")
root=student.getroot()
columns=["id","name","salary"]
out_df=pd.DataFrame(columns=columns)
for node in root:
    s_id=node.find("id").text if node is not None else None
    name=node.find("name").text if node is not None else None
    salary=node.find("salary").text if node is not None else None
    out_df=out_df.append(pd.Series([s_id,name,salary],index=columns),ignore_index=True)
    
out_df
```
