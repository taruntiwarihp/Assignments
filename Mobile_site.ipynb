```python
response=requests.get("https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=b328d897-1e15-4e69-aec9-d804a402643e")
print(response)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
print(soup)
```

```python
title_div=soup.find("div",attrs={"class":"_3wU53n"})
print(title_div)
```

```python
title_divs=soup.find_all("div",attrs={"class":"_3wU53n"})
title=[title.text for title in title_divs]
print(title)
```

```python
card=soup.find("div",attrs={"class":"_1UoZlX"})

title_div=card.find("div",attrs={"class":"_3wU53n"})
title_txt=title_div.text

price_div=card.find("div",attrs={"class":"_2rQ-NK"})
price_txt=price_div.text

rating_div=card.find("div",attrs={"class":"hGSR34"})
rating_txt=rating_div.text

detail_page_link=card.find("a",attrs={"class":"_31qSD5"})
home_url="https://www.flipkart.com"
mob_page_url=home_url+detail_page_link.get("href")
# print(mob_page_url)

response2=requests.get(mob_page_url)
# print(response2)

soup2=BeautifulSoup(response2.content,"html.parser")
seller_name=soup2.find("div",attrs={"id":"sellerName"})
seller_txt=seller_name.text

print(f"{[title_txt]} {[price_txt]} {[rating_txt]}* {[seller_txt]}")
```

```python
#12 pages
import requests
from bs4 import BeautifulSoup
cards=soup.find_all("div",attrs={"class":"_1UoZlX"})
count=1
print(f"{count}page***************")
while(True):
    for card in cards:
        title_div=card.find("div",attrs={"class":"_3wU53n"})
        title_txt=title_div.text

        price_div=card.find("div",attrs={"class":"_2rQ-NK"})
        price_txt=price_div.text

        rating_div=card.find("div",attrs={"class":"hGSR34"})
        rating_txt=rating_div.text

        detail_page_link=card.find("a",attrs={"class":"_31qSD5"})
        home_url="https://www.flipkart.com"
        mob_page_url=home_url+detail_page_link.get("href")

        response2=requests.get(mob_page_url)
        
        soup2=BeautifulSoup(response2.content,"html.parser")
        seller_name=soup2.find("div",attrs={"id":"sellerName"})
        seller_txt=seller_name.text

        print(f"{title_txt} {price_txt} {rating_txt}* {seller_txt}")

    
    next_btn_link=soup.find("a",attrs={"class":"_3fVaIS"})
    home_url="https://www.flipkart.com"
    new_url=home_url+next_btn_link.get("href")
    response=requests.get(new_url)
    soup=BeautifulSoup(response.content,"html.parser")
    cards=soup.find_all("div",attrs={"class":"_1UoZlX"})
    count+=1
    print(f"{count}page***************")
    if count>10:
        break
    
print("bye")
```
